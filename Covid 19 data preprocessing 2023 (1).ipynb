```python
covid_df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],inplace=True,axis=1)
```

```python
covid_df.head(1)
```

```python
covid_df['Date']=pd.to_datetime(covid_df['Date'],format='%Y-%m-%d')
```

```python
covid_df.head()
```

```python
covid_df['Active_cases']=covid_df['Confirmed']-(covid_df['Cured']+covid_df['Deaths'])
```

```python
covid_df.tail()
```

```python
statewise=pd.pivot_table(covid_df,values=['Confirmed','Deaths','Cured'],index='State/UnionTerritory',aggfunc=max)
```

```python
statewise
```

```python
covid_df
```

```python
statewise['Recovery Rate']=statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise['Mortality Rate']=statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise=statewise.sort_values(by='Confirmed',ascending=False)
```

```python
statewise.style.background_gradient(cmap='cubehelix')
```

```python
Top10_active_cases=covid_df.groupby(by='State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by=['Active_cases'],ascending=False).reset_index()
```

```python
Top10_active_cases
```

```python
fig=plt.figure(figsize=(16,9))
```

```python
plt.title("top 10 states with most active cases",size=25)
```

```python
ax=sns.barplot(data=Top10_active_cases.iloc[:10],y='Active_cases',x='State/UnionTerritory',linewidth=2,edgecolor='red')
```

```python
Top10_active_cases=covid_df.groupby(by='State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by=['Active_cases'],ascending=False).reset_index()
fig=plt.figure(figsize=(16,9))
plt.title("top 10 states with most active cases",size=25)
ax=sns.barplot(data=Top10_active_cases.iloc[:10],y='Active_cases',x='State/UnionTerritory',linewidth=2,edgecolor='red')
```

```python
#Top 10 states with highest death
Top10_deaths=covid_df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()
fig=plt.figure(figsize=(16,9))
plt.title("top 10 states with most death",size=25)
ax=sns.barplot(data=Top10_deaths.iloc[:12],y='Deaths',x='State/UnionTerritory',linewidth=2,edgecolor='red')
```

```python
fig=plt.figure(figsize=(12,6))
ax=sns.lineplot(data=covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Kerala','Uttar Pradesh','Tamil Nadu'])],
 x='Date',y='Active_cases',hue='State/UnionTerritory') 
ax.set_title("top 5 affected city in india",size=16)
```

```python
#Remove rows where state=india
vaccine=vaccine_df[vaccine_df.State!='India']
```

```python
vaccine
```

```python
vaccine.rename(columns={'Total Individuals Vaccinated':'Total'},inplace=True)
```

```python
vaccine
```

```python
#Most vaccinated state
max_vac=vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac=max_vac.sort_values('Total',ascending=False)[:5]
max_vac
```

```python

fig=plt.figure(figsize=(16,9))
plt.title("top 5 vaccinated states in india",size=25)
x=sns.barplot(data=max_vac.iloc[:10],y=max_vac.Total,x=max_vac.index,linewidth=2,edgecolor='red')
plt.show()
```
